```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import glob
```

```python
import re
img_path_list = glob.glob("./image/valu_cutout_face_over64/*.jpg")
img_path_list.sort()
values = []
for i, img_path in enumerate(img_path_list):
        img_file = img_path.replace("./image/valu_cutout_face_over64/", "")
        img_file = re.sub("__.+__", "", img_file)
        img_file = re.sub(".jpg", "", img_file)
        img_file = img_file.replace(",", "")
        img_file = round(float(img_file), 3)
        #img_file = str(img_file)
        values.append(img_file)
values.sort()
print(values)
```

```python
# generate data
x = values

plt.hist(x, normed=True)

# プロット表示(設定の反映)
plt.show()
```
